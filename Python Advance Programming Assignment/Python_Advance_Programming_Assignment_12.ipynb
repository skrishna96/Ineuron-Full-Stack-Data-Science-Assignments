### 1. For this challenge, forget how to add two numbers together. The best explanation on what to do for this function is this meme:

#### Examples
#### meme_sum(26, 39)  515
#### # 2+3 = 5, 6+9 = 15
#### # 26 + 39 = 515

#### meme_sum(122, 81)  1103
#### # 1+0 = 1, 2+8 = 10, 2+1 = 3
#### # 122 + 81 = 1103

#### meme_sum(1222, 30277)  31499

```python
def meme_sum(num1,num2):
    n1 = str(num1)
    n2 = str(num2)
    if len(n1)> len(n2):
        n2 = n2.rjust(len(n1),'0')
    elif len(n1) < len(n2):
        n1 = n1.rjust(len(n2),'0')
    c = ''
    for i in range(len(n1)):
        c += str(int(n1[i]) + int(n2[i]))
    return int(c)
```

```python
meme_sum(1222, 30277)
```

### 2. Given an integer, create a function that returns the next prime. If the number is prime, return the number itself.

#### Examples
#### next_prime(12)  13
#### next_prime(24)  29
#### next_prime(11)  11
#### # 11 is a prime, so we return the number itself.

```python
def next_prime(num):
    p = 0
    for i in range(2,num):
        if num%i == 0:
            p = 0
            num = num + 1
            break
        else:
            p = 1
    if p == 1:
        return num
    else:
        return next_prime(num)
```

```python
next_prime(12)
```

```python
next_prime(24)
```

```python
next_prime(11)
```

### 3. If a person traveled up a hill for 18mins at 20mph and then traveled back down the same path at 60mph then their average speed traveled was 30mph.

### Write a function that returns the average speed traveled given an uphill time, uphill rate and a downhill rate. Uphill time is given in minutes. Return the rate as an integer (mph). No rounding is necessary.
#### Examples
#### ave_spd(18, 20, 60)  30
#### ave_spd(30, 10, 30)  15
#### ave_spd(30, 8, 24)  12

```python
def ave_spd(minutes,upspeed,downspeed):
    d = upspeed*(minutes / 60)
    downtime = d / downspeed
    return int((2 * d) / ((minutes / 60) + downtime))
```

```python
ave_spd(18, 20, 60)
```

```python
ave_spd(30, 10, 30)
```

```python
ave_spd(30, 8, 24)
```

### 4. The Kempner Function, applied to a composite number, permits to find the smallest integer greater than zero whose factorial is exactly divided by the number.

#### kempner(6)  3
#### 1! = 1 % 6 > 0
#### 2! = 2 % 6 > 0
#### 3! = 6 % 6 === 0

#### kempner(10)  5
#### 1! = 1 % 10 > 0
#### 2! = 2 % 10 > 0
#### 3! = 6 % 10 > 0
#### 4! = 24 % 10 > 0
#### 5! = 120 % 10 === 0
#### A Kempner Function applied to a prime will always return the prime itself.
#### kempner(2)  2
#### kempner(5)  5
#### Given an integer n, implement a Kempner Function.
#### Examples
#### kempner(6)  3
#### kempner(10)  5
#### kempner(2)  2

```python
def factorial(num):
    if num == 1:
        return 1
    else:
        return num * factorial(num - 1)
def kempner(num):
    for i in range(1,num+1):
        if factorial(i) % num == 0:
            return i
```

```python
kempner(6)
```

```python
kempner(10)
```

```python
kempner(2)
```

```python
kempner(5)
```

### 5. You work in a factory, and your job is to take items from a conveyor belt and pack them into boxes. Each box can hold a maximum of 10 kgs. Given a list containing the weight (in kg) of each item, how many boxes would you need to pack all of the items?

#### Example
#### boxes([2, 1, 2, 5, 4, 3, 6, 1, 1, 9, 3, 2])  5
#### # Box 1 = [2, 1, 2, 5] (10kg)
#### # Box 2 = [4, 3] (7kg)
#### # Box 3 = [6, 1, 1] (8kg)
#### # Box 4 = [9] (9kg)
#### # Box 5 = [3, 2] (5kg)

```python
def boxes(lst):
    temp = []
    new_lst = []
    for i in lst:
        if sum(temp) >= 10:
            new_lst.append(temp[:-1])
            a = temp[-1]
            temp = []
            temp.append(a)
            temp.append(i)
        else:
            temp.append(i)
    return f'{new_lst} -> {len(new_lst)}'
```
