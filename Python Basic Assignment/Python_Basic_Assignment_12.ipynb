### 1. In what modes should the PdfFileReader() and PdfFileWriter() File objects will be opened?

__PdfFileReader()__ file objects should be opened in __'rb'__ mode i.e read binary mode.<br>
__PdfFileWriter()__ File objects should be opened in __'wb'__ mode i.e write binary mode.<br>

### 2. From a PdfFileReader object, how do you get a Page object for page 5?

getPage() method is used to get required page.

```python
from PyPDF2 import PdfFileReader
with open('SamplePdf.pdf','rb') as f:
    pdf = PdfFileReader(f)
    print(pdf.getPage(5))
```

### 3. What PdfFileReader variable stores the number of pages in the PDF document?

__getNumPages()__ in __PdfFileReader__ stores the number of pages in the PDF document.

```python
from PyPDF2 import PdfFileReader
with open('SamplePdf.pdf','rb') as f:
    pdf = PdfFileReader(f)
    print(pdf.getNumPages())
```

### 6. What is the difference between a Run object and a Paragraph object?

The structure of sample word document is represented by three different data types in Python-Docx.<br>
At the highest level, a Document object represents the entire document. <br>
The Document object contains a list of Paragraph objects for the paragraphs in the document. (A new paragraph begins whenever the user presses ENTER or RETURN while typing in a Word document.)<br>
Each of these Paragraph objects contains a list of one or more Run objects. A Run object is a contiguous run of text with the same style. A new Run object is needed whenever the text style changes.

```python
import docx
doc = docx.Document('sample.docx')
print("Number of paragraphs: ", len(doc.paragraphs))
print("First paragraph: ",doc.paragraphs[0].text)
print("Number of runs in first paragraph: ",len(doc.paragraphs[0].runs))
print(doc.paragraphs[0].runs[0].text)
print(doc.paragraphs[0].runs[1].text)
print(doc.paragraphs[0].runs[2].text)
print(doc.paragraphs[0].runs[3].text)
```

### 10. How do you create a Document object for a new Word document?

```python
from docx import Document
document = Document()
document.add_paragraph("This is first paragraph")
document.save('sample2.docx')
```

```python
with open('sample2.docx','rb') as f:
    doc = Document(f)
    print(doc.paragraphs[0].text)
```

### 12. What integers represent the levels of headings available in Word documents?

0 to 9 integers represent the levels of headings available in Word documents. To add a title or heading we will use the inbuilt __.add_heading()__ method of the document object.

```python
from docx import Document
doc = Document()
doc.add_heading("Title of document",0)
doc.add_heading("Heading 1",1)
doc.add_heading("Heading 2",2)
doc.add_heading("Heading 3",3)
doc.add_heading("Heading 4",4)
doc.add_heading("Heading 5",5)
doc.add_heading("Heading 6",6)
doc.add_heading("Heading 7",7)
doc.add_heading("Heading 8",8)
doc.add_heading("Heading 9",9)
doc.save('headings.docx')
```
